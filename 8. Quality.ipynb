```
africa = ['South Africa', 'Morocco', 'Kenya', 'Egypt', 'Nigeria']
america = ['United States', 'Canada', 'Uruguay', 'Mexico', 'Ecuador', 'Panama', 'Argentina', 'Brazil', 'Chile', 'Colombia', 'Peru', 'Venezuela']
asia = ['Oman', 'Japan', 'United Arab Emirates', 'Qatar', 'Singapore', 'Saudi Arabia', 'Cyprus', 'Israel', 'Taiwan (China)', 'South Korea', 'Malaysia', 'Kuwait', 'Turkey', 'Georgia', 'India', 'Jordan', 'Azerbaijan', 'China', 'Pakistan', 'Hong Kong (China)', 'Thailand', 'Indonesia', 'Kazakhstan', 'Vietnam', 'Lebanon', 'Philippines', 'Iran', 'Sri Lanka', 'Bangladesh']
europe = ['Netherlands', 'Denmark', 'Switzerland', 'Luxembourg', 'Finland', 'Iceland', 'Austria', 'Norway', 'Germany', 'Sweden', 'Spain', 'Estonia', 'Slovenia', 'United Kingdom', 'Portugal', 'Czech Republic', 'Croatia', 'Lithuania', 'Belgium', 'Ireland', 'France', 'Latvia', 'Slovakia', 'Italy', 'Poland', 'Hungary', 'Romania', 'Bulgaria', 'Bosnia And Herzegovina', 'Greece', 'Serbia', 'Belarus', 'Malta', 'Ukraine', 'North Macedonia', 'Russia']
oceania = ['Australia', 'New Zealand']

afric_dict = dict.fromkeys(africa, 'Africa')
americ_dict = dict.fromkeys(america, 'America')
asia_dict = dict.fromkeys(asia, 'Asia')
europe_dict = dict.fromkeys(europe, 'Europe')
oceania_dict = dict.fromkeys(oceania, 'Oceania')

dictionary = afric_dict | americ_dict | asia_dict | europe_dict | oceania_dict

df_quality['Continent'] = df_quality['Country'].map(dictionary)
df_quality.columns
```
