# Cost of Living

```
output_list = []
df_dict = {}
with open("bestCountry/bestCountry/cost.txt","r") as f:
    output_list = ast.literal_eval(f.read()) #Read data, pass data to new variable

for data_point in output_list:
    df_dict[data_point[0]] = [data[0] if len(data)==1 else "-1" for data in data_point[1:]] #Sort data into respective column

df_cost = pd.DataFrame(df_dict)
columns = df_cost.columns.tolist()
columns[1] = "Country"
df_cost.columns = columns

df_cost = df_cost.drop(['Rank'], axis = 1) #Delete the first column
df_cost.to_csv("data/cost.csv", index = None) # Save data to csv
```
